```python
df = pd.get_dummies(df, columns=['Locality', 'City','Cuisines'])
```

```python
rating_map = {
    'Excellent': 4,
    'Good': 3,
    'Average': 2,
    'Poor': 1,
    'Not Rated': 0
}

df['rating_encoded'] = df['Rating text'].map(rating_map)
```

```python
df.head()
```

```python
df = df.drop(columns=['Rating text'])
```

```python
df.head()
```

```python
df['rating_encoded'] = df['rating_encoded'].fillna(2)
```

```python
df.head()
```

```python
# Select all boolean columns
bool_cols = df.select_dtypes(include='bool').columns

# Convert all boolean columns to integers (True->1, False->0)
df[bool_cols] = df[bool_cols].astype(int)
```

```python
df.head()
```

```python
feature_cols = [col for col in df.columns if df[col].dtype in ['int64', 'float64']]
X = df[feature_cols]

```

```python
import numpy as np

# Initialize a zero vector for all features
user_pref = np.zeros(len(feature_cols))
#set user preferences
user_pref[feature_cols.index('Cuisines_World Cuisine, Mexican, Italian')] = 1
user_pref[feature_cols.index('City_New Delhi')] = 1
user_pref[feature_cols.index('rating_encoded')] = 4
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Calculate similarity scores (user vs all restaurants)
similarity_scores = cosine_similarity([user_pref], X)[0]

# Get indices of top similar restaurants
top_indices = similarity_scores.argsort()[::-1][:5]  # Top 5 recommendations
```
